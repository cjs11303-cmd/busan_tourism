```python
import requests

url = "https://apis.data.go.kr/B551011/KorService2/locationBasedList2"
serviceKey = "ePSdJxdLS6tqJrxStuufUYpC2pFIX0nPUGq94RE45W5ipNMft6nTH94Hj63v5fwjjSYjKTINLuo0uaHOAa6rqg=="

params = {
    'serviceKey': serviceKey,
    'numOfRows': '50',
    'pageNo': '1',
    'MobileOS': 'ETC',
    'MobileApp': 'AppTest',
    '_type': 'json',
    'arrange': 'O',
    'mapX': '129.1585', # 해운대
    'mapY': '35.1587',
    'radius': '10000',
    'contentTypeId': '12'
}

response = requests.get(url, params=params)
data = response.json()

try:
    # 1. 데이터 존재 여부 단계별 확인
    items_container = data.get('response', {}).get('body', {}).get('items')
    
    if items_container and items_container != "":
        items = items_container.get('item')
        
        # 2. 데이터가 1개일 때(dict)와 여러 개일 때(list)를 모두 처리
        if isinstance(items, dict):
            items = [items] # 1개면 리스트로 감싸줌
            
        print(f"--- 수집된 이미지 데이터 정보 ---\n")
        for item in items:
            title = item.get('title')
            img_url = item.get('firstimage')
            
            if img_url:
                print(f"라벨(명칭): {title}")
                print(f"이미지 주소: {img_url}\n")
            else:
                print(f"라벨(명칭): {title} (이미지 없음)\n")
    else:
        print("조회된 결과가 없습니다.")

except Exception as e:
    print(f"데이터 추출 중 에러 발생: {e}")
```
